```python
import numpy as np
import pandas as pd
```

```python
filename = 'output/COLRECT_filtered.pickle'

df = pd.read_pickle(filename)

df.head()
```

```python
# KM plot

from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

srv_time_mon = df.SRV_TIME_MON.values.astype(np.float32)
srv_time_yr = srv_time_mon / 12.0

kmf = KaplanMeierFitter()
kmf.fit(srv_time_yr)

kmf.survival_function_
kmf.median_
kmf.plot()

plt.show()
```

```python
# Histogram
srv_time_yr_df = pd.DataFrame(srv_time_yr)
hist = srv_time_yr_df.hist(bins=20)
```
